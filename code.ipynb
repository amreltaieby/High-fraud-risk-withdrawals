```python
import pandas as pd 
pd.set_option('display.max_columns', None)  #to display all columns in the output
df = pd.read_csv('FraudShield_Banking_Data.csv') # Load the dataset
df.dropna(subset=['Previous_Fraud_Count' , 'Unusual_Time_Transaction' , 'Distance_From_Home' , 'Failed_Transaction_Count'], inplace=True) # Remove rows with missing values in critical columns
suspicious_transactions = df[
    (df['Distance_From_Home'] > 500) |#if too far,that's suspicious
    ((df['Card_Type'] == 'Debit') & (df['Transaction_Amount (in Million)'] > 0.9 * df['Account_Balance (in Million)'])) | #if user withdraws >90% of his balance, that's suspicious
   
    (df['Card_Type'] == 'Credit') & (df['Transaction_Amount (in Million)'] > 2 * df['Account_Balance (in Million)']) |#if user takes a loan for over 2x his balance, he may not pay back
    (df['Daily_Transaction_Count'] > 2 * df['Weekly_Transaction_Count']) #if he makes over 2x transactions in a day than his weekly average, that's suspicious
]

suspicious_transactions.to_csv("Potentially_Suspicious_Transactions.parquet", index=False)
```
